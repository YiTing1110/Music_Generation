````python
def getHTML(url):
    response = requests.get(url, headers = headers)
    if response.status_code != 200:
        print(response.status_code)
    return BeautifulSoup(response.text, 'lxml')
````

````python
# Into Free Midi
def getGenres(freemidi):
    genres = getHTML(freemidi).select(".col-xs-6")[1].select('a[href^="genre-"]')
    genresURL = []
    for genre in genres:
        genresURL.append(genre['href'])
    return genresURL
````

````python
# Into Genres
def getArtists(url):
    artistsURL = []
    for artistHerf in getHTML(url).select(".genre-link-text"):
        for artist in artistHerf.select('a[href^="artist-"]'):
            artistsURL.append(artist['href'])
    return artistsURL
````

````python
def getPage(url):
    html = getHTML(url)
    pagesURL = []
    for page in html.select_one(".pagination").select('a[href^="artist"]'):
        pagesURL.append(page['href'])
    return pagesURL
````

````python
# Into Artists
def getMusic(url):
    midisURL = []
    html = getHTML(url)
    for cellHerf in html.select(".artist-song-cell"):
        for cell in cellHerf.select('a[href^="download"]'):
            midisURL.append(cell['href'])
    return midisURL
````

````python
def getMidi(url, genreDir):
    options = webdriver.ChromeOptions()
    options.binary_location = 'C:/Program Files/BraveSoftware/Brave-Browser/Application/brave.exe'
    options.add_argument('--disable-blink-features=AutomationControlled')
    prefs = {
    "download.default_directory": os.path.abspath(genreDir),
    "download.prompt_for_download": False,
    "download.directory_upgrade": True,
    "safebrowsing.enabled": True
    }
    options.add_experimental_option("prefs", prefs)
    browser = webdriver.Chrome(options = options)
    browser.minimize_window()
    try:
        browser.get(url)
        download_button = WebDriverWait(browser, 5).until(EC.presence_of_element_located((By.ID, "downloadmidi")))
        time.sleep(random.random() * random.randint(1, 3))
        download_button.click()
    except:
        print(url)
    time.sleep(5)
    browser.quit()
````

````python
def main():
    for genreURL in getGenres(freemidi)[0:1]:
        os.makedirs(genreURL[6:], exist_ok = True)
        print(genreURL)
        for artistURL in getArtists(freemidi + genreURL):
            print(artistURL)
            for midiURL in getMusic(freemidi + artistURL):
                getMidi(freemidi + midiURL, genreURL[6:])
                time.sleep(random.randint(5, 25))
            for pageURL in getPage(freemidi + artistURL):
                for midiURL in getMusic(freemidi + pageURL):
                    getMidi(freemidi + midiURL, genreURL[6:])
                    time.sleep(random.randint(5, 25))
````
