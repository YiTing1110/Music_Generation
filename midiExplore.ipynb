'offsetSeconds': 音符的起始時間（單位：秒）。

'durationSeconds': 音符的時值（單位：秒）。

'endTimeSeconds': 音符的結束時間（單位：秒）。

'element': 音符對象本身，可以獲取該音符的其他信息，例如音高、音量等。

'voiceIndex': 音符所在的聲部索引。

Lead: 主旋律樂器，通常是樂曲中最突出的旋律樂器，例如主唱或主奏樂器。

Bass: 低音樂器，通常負責演奏樂曲中的低音部分，例如低音結他或低音提琴。

Electric Bass: 電貝斯，是低音結他的電子版本，通常在搖滾、流行等音樂類型中使用。

Rhythm chords: 和弦樂器，通常負責演奏和弦伴奏，例如節奏吉他或鋼琴。

Kick: 重低音樂器，負責演奏樂曲中的節奏部分，例如貝斯鼓或大鼓。

Low Tom 1: 低音筒鼓，通常用於演奏較低音域的鼓聲，例如低音打擊樂器。

Mid Tom 2: 中音筒鼓，通常用於演奏中等音域的鼓聲，例如中音打擊樂器。

Crash 1: 碰撞鈸，通常用於演奏高音域的鈸聲，例如演奏高音部分的鈸。

Crash 2: 碰撞鈸，通常用於演奏高音域的鈸聲，例如演奏高音部分的鈸。
